```python
def degdata(pages):
    data = []
    for page in range(pages):
        result = requests.get("https://es.chaturbate.com/?page="+str(page))
        src = result.content
        soup = BeautifulSoup(src, 'lxml')
        list = soup.find_all(class_="room_list_room")

        for account in range(len(list)):
            test = list[account]
            img = test.find_all("img")[0].attrs['src']
            link = test.find('a').attrs["href"]
            gender = test.find_all("span")[0].attrs["class"][1]
            age = test.find_all("span")[0].get_text()
            country = test.find_all(class_="location")[0].get_text()
            data.append([link,gender,age, country,img])
            
    return data

data = degdata(35)
```
