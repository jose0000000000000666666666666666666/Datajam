### 2. We define some functions.

```python
"""
This functions maps the urls string to a hash generated by the SHA256 
algorithm.
"""
def get_ID(url):
    
    # Initialize hash object.
    url_hash = hashlib.sha256()
    url_hash.update(url.encode())
    url_hash = url_hash.hexdigest()
    
    # Return the Hash.
    return url_hash
```

```python
"""
This function creates a new website and adds it to the global graph, 
using the witness results, and a node ID.
"""
def add_node_to_graph(url, website_info, content_type, score):

    # This indicates python that we will use a global variable defined 
    # previously.
    global graph_
    
    # Extract the Website info.
    website = Website(
        ID = get_ID(url), 
        url = url,
        content_type = content_type,
        language = website_info[0],
        score = score,
        urls = website_info[1],
        social_networks_urls = website_info[2],
        text = website_info[3], 
        names = website_info[4],
        locations = website_info[5],
        iframes = website_info[6],
        organizations = website_info[7], 
        dates = website_info[8],
        phonenumbers = website_info[9], 
        images_urls = website_info[10],
        payment_accounts = website_info[11],
        emails = website_info[12]
    )
    
    # Add website to the graph.
    graph_.add_node(website)
    print("Website added!")
    print(graph_)
    
    # Return website.
    return website
```

```python
"""
Add an edge to a graph.
"""
def add_edge_to_graph(node1, node2):
    
    # This indicates python that we will use a global variable defined 
    # previously.
    global graph_
    
    # Add edge to graph.
    graph_.add_edge(node1, node2)
```

```python
"""
This function gets the plain text of a website.
Soup referes to a Beautiful Soup object.
"""
def get_website_text(soup):
    
    # Call the global session.
    global session
    
    # Call the website.
    for script in soup(["script", "style"]):                   
            script.decompose()
    text = soup.get_text().replace("\n"," ")
    text = re.sub('\s+',' ',text)

    # Return the plain text.
    return text
```

```python
"""
This functions gets a list of urls, and returns the useful ones, 
more precisly, it returns the nodes with interesting text.
"""
def get_info_useful_urls(candidate_urls, treshold=0.6):
    
    # Global variables.
    global explored_nodes
    
    # Winners list.
    winners = list()
    
    # For every possible winner.
    for possible_winner in candidate_urls:
        
        # If node is not explored.
        if get_ID(possible_winner) not in explored_nodes: 
            
            # Add it to list of already explored nodes.
            explored_nodes.append(get_ID(possible_winner))

            # Wait...
            random_time = random.randrange(0, 1)
            time.sleep(random_time)

            # Get text
            soup = getSoupObject(possible_winner)
            text = get_website_text(soup)        
            website_category = cluster_text(text, treshold)
            print(website_category)

            # If the website is interesting...
            if website_category != -1:

                # Compute score.
                score = get_max_probability(text)

                # Add 4-tuples to winners list.
                winners.append([possible_winner, website_category, score, soup])
                
    # Return the winners!
    return winners
```

```python
"""
Explore the graph recursively!
"""
def explore_graph(origin_node, distance):
    
    # Get distance of the last witness from the root.
    global max_distance_from_root
    
    # If the agent is not to far from the root.
    if distance <= max_distance_from_root:
        
        # Get useful nodes.
        useful_urls = get_info_useful_urls(origin_node.getUrls())
        
        # For every useful url.
        for url, content_type, score, soup in useful_urls:
            
            # Wait..
            random_time = random.randrange(1, 3)
            time.sleep(random_time)
            
            # Instanciate a witness.
            witness = TheWitness()
            
            # Scrap the url!
            result = witness.scrap(url, soup)
            
            # Add new node to graph.
            new_node = add_node_to_graph(url, result, content_type, score)
                        
            # Add node to graph.
            add_edge_to_graph(origin_node, new_node)
            
            # Destroy the witness.
            del witness
            
            # Explore the graph again
            explore_graph(new_node, distance + 1)
```

```python
"""
Explore roots!
"""

# Set roots index.
roots_indices = list()

def explore_from_roots(root_list):
    
    global roots_indices
    
    # We assume that all roots are interesting.
    for root in roots_list: 
        
        # Get root info.
        # This returns a list with the url, website category, score and beautiful soup object.
        print("Root: ", root)
        root_info = get_info_useful_urls([root], treshold = 0.1)[0]
        print("Success root exploration: ", bool(len(root_info) == 4))
        
        
        node_root_pos = len(graph_.nodes) - 1
        roots_indices.append(node_root_pos)
        
        # Print.
        print("Exploring a new root...")
        
        if len(root_info) > 0:
        
            # Get info from the root.
            url = root_info[0]
            content_type = root_info[1]
            score = root_info[2]
            soup = root_info[3]

            # Initialize witness.
            w = TheWitness()
            website_info = w.scrap(url, soup)

            # Get website object.
            website = add_node_to_graph(url, website_info, content_type, score)

            # Explore the graph recursively
            explore_graph(origin_node = website, distance = 1)
            
        random_time = random.randrange(1, 3)
        time.sleep(random_time)
        
    # Connect roots.
    for i in range(len(roots_indices)): 
        try: 
            node1 = list(graph_.nodes)[roots_indices[i]]
            node2 = list(graph_.nodes)[roots_indices[i+1]] 
            add_edge_to_graph(node1, node2)
        except: 
            pass
```
