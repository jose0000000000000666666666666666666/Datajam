# 1. Website Class.
---

Hello! Welcome to the first part of the development of our algorithm **The Witness** developed by Aleph Mind! As you read previosly, in the *readme* file of our project, our goal is to analyze the subgraph of the internet, where the criminal organizations of human traffick might be operating. Before going further, we need a clever way to represent the information of the different websites that are hosted on the net. 

Therefore, we present to you the first class of our algorithm: Website, which will be the class responsable for store and encoding the information related to each website discovered by our algorithm, **The Witness**. 

<img src = "./notebook_assets/website.jpg" width = "300">

```python
# Website Class.
class Website:
    
    #--------------------------------------------------------------------------
    # CONSTRUCTOR
    #--------------------------------------------------------------------------               
    def __init__(self,    
                 ID,                         # ID of the website (We apply SHA256 over the website url)
                 url,                        # URL of the website (this will work as the website identifier or ID).
                 content_type,               # Contet Type of the website.
                 language,                   # Language of the website.
                 score,                      # Score of the website.
                 urls = [],                  # URLS found on the website.
                 social_networks_urls=[],    # Social networks urls.
                 text = "",                  # Text of the website.
                 names = [],                 # Names that appeared on the website.
                 locations = [],             # Locations mentioned on website.
                 iframes = [],               # Iframes code (maps or videos) that appear on the website.
                 organizations = [],         # Organization names.
                 dates = [],                 # Dates on the website.
                 phonenumbers = [],          # Phonenumbers on the website.
                 images_urls = [],           # Images urls.
                 payment_accounts= [],       # Payment Accounts (banks, crytpo).
                 emails = [],                  # Emails on the website.
                ):                
        
        # Initialize private variables.
        self.__ID = ID
        self.__url = url
        self.__content_type = content_type
        self.__language = language
        self.__names =names
        self.__locations = locations
        self.__urls = urls
        self.__phonenumbers = phonenumbers
        self.__iframes = iframes
        self.__images_urls = images_urls
        self.__dates = dates
        self.__payment_accounts=payment_accounts
        self.__emails = emails
        self.__organizations = organizations
        self.__text = text
        self.__social_networks = social_networks_urls
        self.__score = score
        
    #--------------------------------------------------------------------------
    # METHODS
    #--------------------------------------------------------------------------
    
    """
    print(Website) returns the node url.
    """
    def __str__(self):
        return "Node URL " + self.__url
    
    """
    This functions prints the elements of a python list.
    """
    def printList(self, list_):
        try: 
            for element in list_:
                print(element)
        except: 
            pass
    
    """
    This functions prints all the information collected on this website.
    """
    def info(self):
        
        # Website info!
        print("-- Website Information --")
        print("")
        
        # Print website ID.
        print("-- ID --")
        print(self.__ID)
        print("")
        
        # Print website url.
        print("-- URL --")
        print(self.__url)
        print("")
        
        # Print content type of website.
        print("-- Content Type --")
        print(self.__content_type)
        print("")
        
        # Print language.
        print("-- Language --")
        print(self.__language)
        print("")
        
        # Print website score.
        print("-- Score --")
        print(self.__score)
        print("")
        
        # Print urls found.
        print("-- URLS Found --")
        print(self.printList(self.__urls))
        print("")
        
        # Print social networks.
        print("-- Social Networks --")
        print(self.printList(self.__social_networks))
        print("")
        
        # Print names.
        print("-- Names found --")
        print(self.printList(self.__names))
        print("")

        # Print locations.
        print("-- Locations --")
        print(self.printList(self.__locations))
        print("")
        
        # Print iframes found.
        print("-- IFRames --")
        print(self.printList(self.__iframes))
        print("")
        
        # Print organizations.
        print("-- Organizations --") 
        print(self.printList(self.__organizations))
        print("")
        
        # Print dates.
        print("-- Dates --")
        print(self.printList(self.__dates))
        print("")
        
        # Print phone number.
        print("-- Phone numbers --")
        print(self.printList(self.__phonenumbers))
        print("")
        
        # Print list of urls found.
        print("-- Images URLS --")
        print(self.printList(self.__images_urls))
        print("")
        
        # Print payment accounts.
        print("-- Payment Accounts --")
        print(self.printList(self.__payment_accounts))
        print("")

        # Print emails.
        print("-- Emails --")
        print(self.printList(self.__emails))
        print("")

        # Print plain text.
        print("-- Plain Text --")
        print(self.__text)
        print("")
        
    """
    This function save a list in a given file.
    """
    def saveList(self, list_, file):
        try: 
            for element in list_:
                file.write(element)
        except: 
            pass
        
    """
    This functions returns a .txt file with all the information collected on this website.
    """
    def toPlainText(self):
        
        # Open a new file.
        file = open("./datasets/" + self.__ID + ".txt", "w")
        
        # Write URL.
        file.write("W#URL")
        file.write("\n")
        file.write(self.__url)
        file.write("\n")
        
        # Write website content type.
        file.write("W#Content")
        file.write("\n")
        file.write(str(self.__content_type))
        file.write("\n")
        
        # Write Language.
        file.write("W#Language")
        file.write("\n")
        file.write(self.__language)
        file.write("\n")
        
        # Write Score.
        file.write("W#Score")
        file.write("\n")
        file.write(str(self.__score))
        file.write("\n")
    
        # Write urls.
        file.write("W#URLS")
        file.write("\n")
        self.saveList(self.__urls, file)
        file.write("\n")
        
        # Print social networks.
        file.write("W#SocialNetworks")
        file.write("\n")
        self.saveList(self.__social_networks, file)
        file.write("\n")
        
        # Writes names.
        file.write("W#Names")
        file.write("\n")
        self.saveList(self.__names, file)
        file.write("\n")

        # Write locations.
        file.write("W#Locations")
        file.write("\n")
        self.saveList(self.__locations, file)
        file.write("\n")

        # Write Iframes.
        file.write("W#IFRames")
        file.write("\n")
        self.saveList(self.__iframes, file)
        file.write("\n")

        # Write organizations.
        file.write("W#Organizations") 
        file.write("\n")
        self.saveList(self.__organizations, file)
        file.write("\n")
        
        # Write dates.
        file.write("W#Dates")
        file.write("\n")
        self.saveList(self.__dates, file)
        file.write("\n")
        
        # Write Phone Numbers.
        file.write("W#Phonenumbers")
        file.write("\n")
        self.saveList(self.__phonenumbers, file)
        file.write("\n")
       
        # Write images.
        file.write("W#Images URLS")
        file.write("\n")
        self.saveList(self.__images_urls, file)
        file.write("\n")

        # Write payment accounts.
        file.write("W#PaymentAccounts")
        file.write("\n")
        self.saveList(self.__payment_accounts, file)
        file.write("\n")

        # Write emails.
        file.write("W#Emails")
        file.write("\n")
        self.saveList(self.__emails, file)
        file.write("\n")

        # Write text.
        file.write("W#Text")
        file.write("\n")
        aux_text = self.__text.encode("utf-8").decode("ascii", "ignore")
        file.write(aux_text)
        file.write("\n")
        
        # Close the file.
        file.close()
    
    """
    This method recomputes all the attributes of the website, 
    except the ID and URL.
    """
    def explore_again(self):
        
        # Generate a new witness to re-explore the page.
        witness = TheWitness(
                origin_url = self.__url,
                destination_url = self.__url,
                distance_from_root = 0 
        )
        
        # Try to scrap the website again!
        try:         
            
            # Update website values.
            result = witness.scrap()
            self.setContentType(result[2])
            self.setLanguage(result[3])
            self.setUrls(result[4])
            self.setSocialNetworks(result[5])
            self.setText(result[6])
            self.setNames(result[7])
            self.setLocations(result[8])
            self.setIframes(result[9])
            self.setOrganizations(result[10])
            self.setDates(result[11])
            self.setNumbers(result[12])
            self.setImages(result[13])
            self.setPaymentAccounts(result[14])
            self.setEmails(result[15])
            
            # Delete the witness.
            del witness
            
            # Return True.
            return True
        
        except:
            
            # Delete the witness-
            del witness
            
            # Return False.
            return False
        
    #--------------------------------------------------------------------------
    # SET METHODS
    #--------------------------------------------------------------------------
    
    def setContentType(self, new_content_type):
        self.__content_type = new_content_type
    
    def setLanguage(self, newLanguage):
        self.__language = newLanguage
        
    def setNames(self, newNames):
        self.__names = newNames   
        
    def setLocations(self, newLocations):
        self.__locations = newLocations
        
    def setUrls(self, newUrls):
        self.__urls = newUrls
        
    def setNumbers(self, newNumbers):
        self.__numbers = newNumbers
        
    def setIframes(self, newIframes):
        self.__iframes = newIframes
        
    def setImages(self, newImages):
        self.__images = newImages
        
    def setDates(self, newDates):
        self.__dates = newDates
        
    def setPaymentAccounts(self, newPaymentAccounts):
        self.__payment_accounts = newPaymentAccounts
        
    def setEmails(self, newEmails):
        self.__emails = newEmails
        
    def setOrganizations(self, newOrganizations):
        self.__organizations = newOrganizations
        
    def setText(self, newText):
        self.__text = newText
        
    def setSocialNetworks(self, newSocialNetworks):
        self.__social_networks = newSocialNetworks
        
    #--------------------------------------------------------------------------
    # GET METHODS
    #--------------------------------------------------------------------------
    
    def getID(self):
        return self.__ID
    
    def getUrl(self):
        return self.__url
    
    def getContentType(self): 
        return self.__content_type
    
    def getLanguage(self):
        return self.__language
    
    def getNames(self):
        return self.__names
    
    def getLocations(self):
        return self.__locations
    
    def getUrls(self):
        return self.__urls
    
    def getNumbers(self):
        return self.__numbers
    
    def getIframes(self):
        return self.__iframes
    
    def getImages(self):
        return self.__images
    
    def getDates(self):
        return self.__dates
    
    def getPaymentAccounts(self):
        return self.__payment_accounts
    
    def getEmails(self):
        return self.__emails
    
    def getOrganizations(self):
        return self.__organizations
    
    def getText(self):
        return self.__text
    
    def getSocialNetworks(self):
        return self.__social_networks
```
