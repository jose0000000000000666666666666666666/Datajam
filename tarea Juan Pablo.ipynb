```python
page_url = "https://streamate.com/"
result = requests.get(page_url)
src = result.content
soup = BeautifulSoup(src, 'html.parser')
soup
```

```python
def get_text():
    global soup, text
    for script in soup(["script", "style"]):                   
        script.decompose()
    text = soup.get_text().replace("\n"," ")
    text = re.sub('\s+',' ',text)
    return text
```
